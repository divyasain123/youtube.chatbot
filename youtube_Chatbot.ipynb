```python
from youtube_transcript_api import YouTubeTranscriptApi

video_id = "UrsmFxEIp5k"

try:
    api = YouTubeTranscriptApi()

    transcript = api.fetch(video_id)

    text = " ".join(chunk.text for chunk in transcript)

    print(text)

except Exception as e:
    print("Error:", e)
```

# **Step 1b - Indexing (Text Splitting)**

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
chunks=splitter.create_documents([text])
```

#Step 2 Retrieval


```python
retriever=vectorstore.as_retriever(search_type="similarity",search_kwargs={"k":4})
```

```python
from langchain_core.prompts import PromptTemplate
```

```python
import os
from langchain_huggingface import ChatHuggingFace, HuggingFaceEndpoint

# Step 1: Define the endpoint
llm = HuggingFaceEndpoint(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    provider="featherless-ai",
    task="conversational",
    huggingfacehub_api_token=os.environ["HUGGINGFACEHUB_API_TOKEN"],  # ← pulls from env
)
# Step 2: Wrap it with ChatHuggingFace
chat_model = ChatHuggingFace(llm=llm)

# Step 3: Invoke using a message format
from langchain_core.messages import HumanMessage
```

```python

prompt=PromptTemplate(
  template="""you are a helpfull assistant.
  Answer ONLY from the provided transcript context.
  If the context is insufficent just say U dont know.

  {context}

  Question: {question}
  """,
  input_variables=["context","question"]
)
```

```python
question      =   "did we discussed anything related to oops in this video?if yes then what "
retriever_docs=retriever.invoke(question)
```

```python
content_text="\n\n".join(doc.page_content for doc in retriever_docs)
content_text
```

```python
final_prompt=prompt.invoke({"context":content_text,"question":question})
```

#Step 4 Genration

```python
# Step 4: Generation
from langchain_core.messages import HumanMessage

# Convert the prompt to a plain string first, then wrap in HumanMessage
answer = chat_model.invoke([HumanMessage(content=final_prompt.text)])
print(answer.content)
```

```python
from langchain_core.runnables import RunnableParallel,RunnablePassthrough,RunnableLambda
from langchain_core.output_parsers import StrOutputParser

```

```python
parser=StrOutputParser()

def format_docs(retriever_docs):
  content_text="\n\n".join(doc.page_content for doc in retriever_docs)
  return content_text
```

```python
parallel_chain=RunnableParallel({
    'context':retriever|RunnableLambda(format_docs),
    'question':RunnablePassthrough()
})
```

```python
parallel_chain.invoke('what is a error handling')
```
