#Custom tool

```python
from langchain.tools import tool
```

```python
#step 1 create a function
def multi(a,b):
  "multipying two numbers"
  return a*b
```

```python
#step 2 add type hint
def multi(a:int,b:int)->int:
  "multipying two numbers"
  return a*b
```

```python
#step 3 add tool decorator
@tool
def multi(a:int,b:int)->int:
  "multipying two numbers"
  return a*b
```

```python
result=multi.invoke({"a":2,"b":3})
print(result)
```

```python
print(multi.args)
print(multi.name)
print(multi.description)
```
